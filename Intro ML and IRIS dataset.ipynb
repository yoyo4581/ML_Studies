

**Training Dataset:** Data inputs. The nature of the data input varies depending on the model used. Usually accompanied by labels in supervised learning that represent the output, unlabeled data in unsupervised learning; and a defined measure of reward for a particular action in reinforcement learning. The data is passed to the machine learning algorithm to train and optimize a predictive model. 

**Testing Dataset:** Half of the data set is used at the very end to evaluate the model. Measuring it's performance to real-world data.

**Validation Dataset:** Used for tuning of the model. In cross-validation, the training dataset is partitioned into the training dataset and the validation dataset. During training, the ML algorithm used leads itself to parameters that are incorporated into the model based on the training data. To determine which ML algorithm is most suitable, and to fine tune that algorithm's parameters by incorporating hyper-parameters to improve performance; we must repeatedly measure the performance of our model independently from the final testing. Here the validation dataset allows us to configure our model prior to its final evaluation.


***
**Ground Truth**: information assigned to the dataset that is provided as factual empirical evidence as opposed to information that is implied. We use this evidence to train or validate our model. In the snail mail example, the labeled dataset served as the ground truth.
    
**Label**: Expected output of a data point. For example, in the snail mail problem, image data is labeled by subject-matter experts categorically. The image data is representative of this value, and patterns in this data should inform a similar outcome. The label is synonymous with the target value.

***

**pre-processing:** fixing data-set structure for the performance of the learning algorithm. Processing (substantiating into values) RAW data features into data that can be interpretable. Can also include steps for scaling data through transformation or standardized distribution to an appropriate range. Features that share a high correlation are subject to dimensionality reduction to lower storage space and increase efficiency. While a smaller feature-space leads to information loss to a certain extent, in certain cases it can benefit the predictive power of the model if the data set has a significant amount of features with poor data quality.

**feature:** measurements, or attributes of a datapoint, or dimensions of dataset. Refering to columns in dataset, the values that belong to an individual data point. In the Iris dataset this is the measurements of the pedals, in the snail mail dataset this is pixel values (darkness) belonging in the 8x8 image.

**numerical:** data that is a number that or a measurable value.

**nominal:** data that is classified without order.


***

**Decision Surface:** How classification algorithm decides what input feature is defined into what classification. Decision surfaces are defined by functions; SVM chooses a boundary that maximizes the seperation between two classes, it solves for the optimum solution. Artificial Neural Networks place the boudaries to minimize local classification error via gradient descent optimization algorithm.

***

**Model Validation:** A technique to confirm that a particular trained model with certain hyper-parameters achieves its purpose by evaluating its predictive performance against a validation dataset. 

**Accuracy:** a typical measure used to evaluate a model. The ratio of correct predictions over the ratio of total predictions. Can be represented through a confusion matrix, where each column shows the predicted class (the guess), and each row shows the expected class (what it should be).

**Cross-validation:** Technique to address the question of deterimining which model performs best on the test dataset if we don't use this test dataset for model selection. In model selection we are adjusting different parameters and comparing different models to make better predictions.

To summarize the problem, if we are to reuse the same test dataset for model selection, our test dataset becomes part of our training data, leading to an overfit model. One of the ways to offset this is through cross-validation. Hold-out cross-validation is one type; using a validation dataset to repeatedly evaluate the performace of the model after training with different hyperparameters. Once satisfied we estimate the models performance on the test data. However this method is highly dependent on the partitioning of the training dataset into the training and validation dataset.

Another cross-validation technique is k-fold cross-validation. Think of this as resampling. Instead of two partitions of the training dataset, into the training and validation dataset. We make k partitions of the training dataset. All but one partition or fold is used for training, that one fold is used for validation. This is repeated k times, picking a different dataset fold for validation; to obtain k models and performance estimates allowing for an average performance score. This estimate is less sensitive to partitioning, and when optimal hyperparameter values are selected, the model is then retrained on the complete training dataset and tested on the testing dataset.

***
**Parameters:** variables in different learning algorithms that are set during the learning of our model. These are typically default and may not be optimized for our specific problem.

**Hyperparameters:** Parameters not learned from input data that allow the adjustment of the model for increased performance.

**Overfit:** Model overfitting is a learning behavior where the model is a good representation of the training data, but performs poorly for new data. The model lacks generalizeability due to a variety of reasons. Training data is too small, training data contains noisy data, model trains too long on a single sample set of data, and the model complexity is overly high (sensitive to noise within the training dataset).
```python
# Standard scientific Python imports
%matplotlib inline
import matplotlib.pyplot as plt
plt.rcParams["figure.dpi"] = 72
import numpy as np

# Import datasets, classifiers and performance metrics
from sklearn import datasets
```

```python
# The digits dataset for training
iris = datasets.load_iris()
```

```python
#preprocessing
n_samples = len(iris.data)
X = iris.data.reshape((n_samples, -1))
y = iris.target

print(f'N={len(X)}, M={len(X[0])}')
```

```python
from sklearn import naive_bayes, metrics
from sklearn.model_selection import train_test_split

train_data, test_data, train_labels, test_labels = train_test_split(X, y, test_size=0.5, random_state=42)

# Create a classifier: a Gaussian Naive Bayes
classifier = naive_bayes.GaussianNB(priors=None)

# Learn the digits on the first half of the digits - 50% data is used as the training set
classifier.fit(train_data, train_labels)

# Predict the value of the digit on the second half
expected = test_labels
predicted = classifier.predict(test_data)
```

```python
%%time

acc = metrics.accuracy_score(expected, predicted)

print(f'Classification report for classifier {classifier}:\nAccuracy={acc:.3f}\n')
print(f'Confusion matrix:\n{metrics.confusion_matrix(expected, predicted)}')
```

```python
from sklearn.tree import DecisionTreeClassifier

clf = DecisionTreeClassifier(random_state = 0)
clf.fit(train_data, train_labels)

expected = test_labels
predicted = clf.predict(test_data)
```

```python
%%time

acc = metrics.accuracy_score(expected, predicted)

print(f'Classification report for classifier {clf}:\nAccuracy={acc:.3f}\n')
print(f'Confusion matrix:\n{metrics.confusion_matrix(expected, predicted)}')
```

Now I will perform training with increasing proportion of the dataset. I will store the accuracy of the model after each training step, then plot.

```python
from sklearn import datasets, naive_bayes, metrics
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
import numpy as np
import matplotlib.pyplot as plt
from statistics import mean, pstdev

test_prop = np.arange(0.95, 0.15, -0.05)

iris = datasets.load_iris()
n_samples = len(iris.data)
X = iris.data.reshape((n_samples, -1))
y = iris.target

acclist_bayes = list()
accstdev_bayes = list()
acclist_tree = list()
accstdev_tree = list()

for i in test_prop:
    # Create a classifier: a Gaussian Naive Bayes
    acc_tree = list()
    acc_bayes = list()
    
    classifier = naive_bayes.GaussianNB(priors=None)
    clf = DecisionTreeClassifier(random_state = 0)
    
    for v in range(10):
        train_data, test_data, train_labels, test_labels = train_test_split(X, y, test_size=i)
        

        # Learn the digits on the first half of the digits
        classifier.fit(train_data, train_labels)
        # Predict the value of the digit on the second half
        expected_bayes = test_labels
        predicted_bayes = classifier.predict(test_data)
        acc_bayes.append(metrics.accuracy_score(expected_bayes, predicted_bayes))


        clf.fit(train_data, train_labels)
        expected_tree = test_labels
        predicted_tree = clf.predict(test_data)

        acc_tree.append(metrics.accuracy_score(expected_tree, predicted_tree))
    
    acclist_bayes.append(mean(acc_bayes))
    accstdev_bayes.append(pstdev(acc_bayes))
    
    acclist_tree.append(mean(acc_tree))
    accstdev_tree.append(pstdev(acc_tree))


plt.errorbar(test_prop, acclist_bayes, accstdev_bayes, capsize = 0.5)
plt.xlabel('Test proportion')
plt.ylabel('Accuracy')

plt.errorbar(test_prop, acclist_tree, accstdev_tree, capsize = 0.5)
plt.xlabel('Test proportion')
plt.ylabel('Average Accuracy')
plt.legend(['Gausian Bayes', 'Decision Tree'] )
plt.savefig('Average Accuracy of Classifiers with Stdev')
plt.show()


print('Test_prop', test_prop, '\n')
print('Accuracy Bayes', acclist_bayes, '\n','Error Bayes', accstdev_bayes, '\n')
print('Accuracy Tree', acclist_tree, '\n', 'Error Tree', accstdev_tree, '\n')
```
